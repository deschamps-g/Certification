```python
import pandas as pd
data_analysis = pd.read_excel('https://full-stack-assets.s3.eu-west-3.amazonaws.com/Deployment/get_around_delay_analysis.xlsx')
data_analysis.head()
```

```python
def generic_describe(df):

    print('Number of rows:',len(df))

    print('First few lines:')
    display(df.head())

    print('Basic statistics:')
    display(df.describe(include="all"))

    print('Missing values:')
    display(df.isnull().sum()/len(df)*100)

generic_describe(data_analysis)
```

```python
import plotly.express as px

fig = px.histogram(data_analysis,x='checkin_type',color='state')
fig.show()
```

```python
# percentage of canceled orders

data_analysis[data_analysis['state']=='canceled'].count().iloc[0]/len(data_analysis)*100
```

```python
# percentage of canceled mobile orders

checkin_type = 'mobile'
data_mobile = data_analysis[data_analysis['checkin_type']==checkin_type]
data_mobile[data_mobile['state']=='canceled'].count().iloc[0]/len(data_mobile)*100
```

```python
# percentage of canceled connect orders

checkin_type = 'connect'
data_connect = data_analysis[data_analysis['checkin_type']==checkin_type]
data_connect[data_connect['state']=='canceled'].count().iloc[0]/len(data_connect)*100
```

```python
# we use ~ to get the negation of what is indicated

data_analysis[~data_analysis['delay_at_checkout_in_minutes'].isna()]
```

```python
data = data_analysis[~data_analysis['delay_at_checkout_in_minutes'].isna()]
fig = px.histogram(data,color='checkin_type', x='delay_at_checkout_in_minutes')
fig.show()
```

```python
col = 'delay_at_checkout_in_minutes'
col_mean = data[data[col] > 0][col].std()
print(col_mean)
```

```python
col = 'delay_at_checkout_in_minutes'

col_mean = data[col].mean()
col_med = data[col].median()
col_std = data[col].std()
lower_bound = 0   # we only look at people who are late
upper_bound = col_mean + 0.5 * col_std
df = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
display(df.head())
print(f'mean: {col_mean}  median: {col_med}  std: {col_std}   len: {len(df)}')
```

```python
fig = px.histogram(df,color='checkin_type', x='delay_at_checkout_in_minutes')
fig.show()
```

```python
threshold = 500
print(f'Percentage of cars over the threshold : {len(data[(data[col] > threshold)])/len(data) * 100:.2f}%')
```

```python
percent_of_car_late = 0 
threshold = 500
while percent_of_car_late <5 :
    percent_of_car_late = len(data[(data[col] > threshold)])/len(data) * 100
    threshold-=1
print(threshold)
print(f'Part of cars over the threshold: {percent_of_car_late:.2f}%  or {int(percent_of_car_late*len(data))} cars')
```

```python
def car_late(data):
    return len(data[(data[col] > threshold)])/len(data) * 100
```

```python

percent_of_car_late_mobile = car_late(data_mobile)
print(f'Percent of mobile cars over the threshold : {percent_of_car_late_mobile:.2f}%  or {int(percent_of_car_late_mobile*len(data))} cars')
percent_of_car_late_connect = car_late(data_connect)
print(f'Percent of connect cars over the threshold : {percent_of_car_late_connect:.2f}%  so {int(percent_of_car_late_connect*len(data))} cars')
```
